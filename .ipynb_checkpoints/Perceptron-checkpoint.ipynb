```python
import random
import numpy as np
```

```python
def derivada(x):
    return x * (1 - x)
```

```python
x = 0.85
y = 0.58
w = random.random()
```

```python
for i in range(10):
    a = np.tanh(x * w)
    
    e = y - a
    
    w += x * derivada(e)
    
    print(a)
```
